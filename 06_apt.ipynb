## Tool Information

- **Tool:** AcinetobacterPlasmidTyping (Hamidian et al.)

- **Organism:** *Acinetobacter baumannii*
- **Analysis type:** Plasmid replicon typing

This notebook documents the identification of plasmid replicon types in *Acinetobacter baumannii* genomes using the AcinetobacterPlasmidTyping marker database developed by Hamidian *et al.*

The AcinetobacterPlasmidTyping database consists of curated replication initiation (rep) gene sequences representing the major *Acinetobacter* plasmid families, including R1, R3, and RP groups. These markers enable classification of plasmids based on replicon homology.

The analysis was performed on assembled genome contigs generated from the GHRU assembly pipeline. Identified plasmid replicon types were used for downstream plasmid epidemiology analysis and integration with antimicrobial resistance and mobile genetic element profiling.

## Execution

Plasmid replicon typing was performed using BLASTn by screening genome assemblies against the AcinetobacterPlasmidTyping marker database.

```python
%%bash
cd /home/nidhi/tools/AcinetobacterPlasmidTyping

mkdir -p /data/internship_data/nidhi/aba/new_output/apt_output

for f in /data/internship_data/nidhi/aba/new_output/nextflow_output/assemblies/*.fasta; do
    sample=$(basename $f .fasta)

    blastn \
      -query marker_db/acinetobacterplasmidtyper.fasta \
      -subject $f \
      -outfmt 6 \
      -perc_identity 95 \
      -out /data/internship_data/nidhi/aba/new_output/apt_output/${sample}.tsv
done
```
